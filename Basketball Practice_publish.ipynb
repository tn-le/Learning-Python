It's not updated data to reflect the current trends of this kind of game but the purpose of this exercise is to be getting familiar with using Python to look at the insights of datasets.  
To my humble understanding about NBA, an NBA regular season has 82 games which occur from October to April. Each game has four 12-minute quarters, meaning 48 minutes per game. According to NBA, "A legal field goal or free throw attempt shall be scored when a ball from the playing area enters the basket from above and remains in or passes through the net. A successful field goal attempt from the area on or inside the three-point field goal line shall count two points. A successful field goal attempt from the area outside the three-point field goal line shall count three points". A successful Field Goal counts either 2 points or 3 points depending on the position where the player attempted that field goal that would build up the player's style. (We can understand that the successful field goal attempt made in the area further away the basket shall count 3 points, and the closer to the basket shall count 2 points). A successful Free Throw counts 1 point (Free Throw could be understood as a Penalty in football).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
plt.rcParams['figure.figsize'] = 20, 8
```

```python
def myplot(data, playerlist = Players):
    color = {"KobeBryant":"black","JoeJohnson":"Red","LeBronJames":"Green","CarmeloAnthony":"Blue","DwightHoward":"Yellow","ChrisBosh":"Pink","ChrisPaul":"Orange","KevinDurant":"Brown","DerrickRose":"Purple","DwayneWade":"Magenta"}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=color[name], ls="--", marker="s", ms=7, label=name)
    plt.legend()
    plt.xticks(list(range(0,10)),Seasons, rotation='vertical')
    plt.show()
```

```python
# Salaries of all players over the seasons
myplot(Salary)
```

Kober Bryant's salary is always much higher than his colleagues in almost seasons.

The most noticable insight from two plots above is DwightHoward's free throw accuracy is extremely poor compared to others. However, his field goal accuracy was exceptional. Let's uncover more to see each player's playing style to see if a player was a 2-point or 3-point maker.

```python
#Player Style, excluding Free Throw:
#Because the Points matrix includes all types of point: 1 point, 2 points and 3 points.
#So, we want to get out 1 point type to get a new matrix including only 2 and 3 points.
# As we know, each free throw worths 1 point, we get Points minus FreeThrows
TwoAndThreePointMatrix = Points - FreeThrows
TwoAndThreePointMatrix
```

```python
# How many points per field goal?
myplot(TwoAndThreePointMatrix/FieldGoals)
```

Generally, all players changed their playing style over the seasons except Dwight Howard who was always the 2-point maker. That's also the reason why his field goal accuracy was the best. It means he can score in the area close to the basket or even in contact with it.
